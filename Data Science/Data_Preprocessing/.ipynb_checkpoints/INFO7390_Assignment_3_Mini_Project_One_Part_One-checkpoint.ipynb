###   1.4.4 Finding Missing Values <a id='finding_missing_values'></a>

Missing Data can occur when no information is provided for one or more items or for a whole unit. Missing Data is a very big problem in real life scenario. Missing Data can also refer to as NA(Not Available) or Null values in pandas. In DataFrame sometimes many datasets simply arrive with missing data, either because it exists and was not collected or it never existed. For Example, Suppose different user being surveyed may choose not to share their income, some user may choose not to share the address in this way many datasets went missing.
It is very important to check for these missing values and find their location.

[Back to top](#Introduction)

```python
print("Are there Null Values in the dataset? ")
df_1.isnull().values.any()
```

As suspected, there are null values in the dataset, next we find the total percentage of these null values in our dataset.

```python
missing_values_count = df_1.isnull().sum()
# how many total missing values do we have?
total_cells = np.product(df_1.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100

print("Total percent of missing data: {} %".format (percent_missing))
```

Now that we know that 4.64% of our dataset is null, it doesn't seem like a lot, but that might not be true. 4% in the whole dataset might be a relatively small value, but we also need to check the distribution of these null values in the different columns, which we do below

```python
# df_1.isnull().sum()
total = df_1.isnull().sum()[df_1.isnull().sum() != 0].sort_values(ascending = False)
percent = pd.Series(round(total/len(df_1)*100,2))
pd.concat([total, percent], axis=1, keys=['Total Missing', 'Percent'])
```

As suspected, couple of columns like BuildingArea and YearBuilt have as much as 47% & 39% of data missing, which is quite significant.

### 1.4.5 Figure out Why the Data is Missing <a id='why_data_missing'></a>

This is the point at which we get into the part of data science that I like to call "data intution", by which I mean "really looking at your data and trying to figure out why it is the way it is and how that will affect your analysis". It can be a frustrating part of data science, especially if you're newer to the field and don't have a lot of experience. For dealing with missing values, you'll need to use your intution to figure out why the value is missing. One of the most important questions you can ask yourself to help figure this out is this:

   >**Is this value missing because it wasn't recorded or because it doesn't exist?**
   
If a value is missing becuase it doesn't exist (like the height of the oldest child of someone who doesn't have any children) then it doesn't make sense to try and guess what it might be. These values you probably do want to keep as NaN. On the other hand, if a value is missing because it wasn't recorded, then you can try to guess what it might have been based on the other values in that column and row. This is called imputation, and we'll learn how to do it next! :)


> **Tip:** This is a great place to read over the dataset documentation if you haven't already! If you're working with a dataset that you've gotten from another person, you can also try reaching out to them to get more information.

If you're doing very careful data analysis, this is the point at which you'd look at each column individually to figure out the best strategy for filling those missing values. Next, we'll cover some "quick and dirty" techniques that can help you with missing values but will probably also end up removing some useful information or adding some noise to your data.

[Back to top](#Introduction)

###  1.4.6 Handling the Missing Values  <a id='handling_missing_values'></a>

Next we will try to handle these missing values in our dataset. As the missing data might just be the key to a perfect ML algorithm.

There are many ways data can end up with missing values. For example
- A 2 bedroom house wouldn't include an answer for how large is the third bedroom
- Someone being surveyed may choose not to share their income
- Python libraries represent missing numbers as nan which is short for "not a number". 

Most machine learning libraries (including scikit-learn) give an error if you try to build a model using data with missing values. So you'll need to choose a strategy to handle these missing values.

![title](Images\Missing_Values.jpg)

[Back to top](#Introduction)

```python
df_1_NN=df_1
df_1_NN=df_1_NN.fillna(0)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(23,20))
plt.subplots_adjust(hspace=0.4)

z = pd.Series()
for col in df_1_NN.columns.values[2:]:
    if ((col!='Type')&(col!='Method')&(col!='SellerG')&(col!='Date')&(col!='CouncilArea')&(col!='Regionname')&(col!='Rooms')&(col!='Price')&(col!='Distance')&(col!='Postcode')&(col!='Bedroom2')&(col!='Bathroom')):
        colums=np.array(df_1_NN[col])
        z[col]=colums
for i in range(2):
    for j in range(2):
        
        x=z[i*2+j]
        y=z.index[i*2+j]
        sns.distplot(x,axlabel=y,ax=axes[i,j])
        
fig.suptitle('Distplot of all the numeric columns of the dataset',fontsize=30)        
plt.show()
```

We can see from the plot that different numerical variables are of different types, some are categorical and some are continuous. We need to know the difefrent types of variables before we do data imputation. We can use one of the three modes of central tendency to fill in the missing values.

**Which mode of central tendency we are using to fill in missing values?**

Reasoning:- By looking at both plot we can say that Car is a categorical variable, Building Area and LandSize are continuous variables with exponential distributions(right skew with 0 being set where value is unknown) and Year Built is a bi-modal distribution (because the rows where they don't have data about the year it was built, it is set to 0)

- For the distributions that are generally skewed and not normal we use **median** as a measure of central tendency.
- **Mean** is not used because it is greatly affected by outliers which would may result in mean being skewed between outliers but median retains its position and is not as strongly influenced by the skewed values. 
- And as for **mode**, it is generally used for categorical data . Also in mode's case , two or more datapoints have same frequency, as in mode may have more than one value. So, we use median to fill in our missing values.






### 1.4.6.1 Drop missing values (Most Dirty Approach) <a id='drop_missing'></a>

If you're in a hurry or don't have a reason to figure out why your values are missing, one option you have is to just remove any rows or columns that contain missing values. (Note: It is not generally recommend to use this approch for important projects! It's usually worth it to take the time to go through your data and really look at all the columns with missing values one-by-one to really get to know your dataset.)

If you're sure you want to drop rows with missing values, pandas does have a handy function, dropna() to help you do this

[Back to top](#Introduction)

```python
df_1_Drop=df_1.dropna()
df_1_Drop.info()
```

Looks like we've deleted half of our dataset while trying to drop null values. This is because over half of the rows in our dataset had at least one missing value. Clearly this is not a very good technique as we are always looking to add to our sample size, not reduce it! We might have better luck removing all the columns that have at least one missing value instead.

### 1.4.6.4 Better Option: Imputation <a id='imputation'></a>

Imputation fills in the missing values with some number. For instance, we can fill in the mean value along each column.

![title](Images\Data_Imp_2.png)

The imputed value won't be exactly right in most cases, but it usually leads to more accurate models than you would get from dropping the column entirely.

The default behavior fills in the mean value for imputation. Statisticians have researched more complex strategies, but those complex strategies typically give no benefit once you plug the results into sophisticated machine learning models.
One (of many) nice things about Imputation is that it can be included in a scikit-learn Pipeline. Pipelines simplify model building, model validation and model deployment.

[Back to top](#Introduction)

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()
data_with_imputed_values = my_imputer.fit_transform(df_1.values)
```

Uh-OH!, looks like one of the variables with null is a categorical variable, and obviuously we can't use a mean or median strategy there. For this we have to use encoding (Covered in part 4). Here we use a cleaver technique where we just impute the data with numerical values

```python
def split_num_categ(df_1):
    categorical_columns = []
    numeric_columns = []
    for c in df_1.columns:
        if df_1[c].map(type).eq(str).any(): #check if there are any strings in column
            categorical_columns.append(c)
        else:
            numeric_columns.append(c)

    #create two DataFrames, one for each data type
    data_numeric = df_1[numeric_columns]
    data_categorical = pd.DataFrame(df_1[categorical_columns])
    return data_numeric,data_categorical

data_numeric,data_categorical=split_num_categ(df_1)

from sklearn.preprocessing import Imputer
imp = Imputer(missing_values=np.nan, strategy='mean', axis=0)
data_numeric = pd.DataFrame(imp.fit_transform(data_numeric), columns = data_numeric.columns) #only apply imputer to numeric columns


#you could do something like one-hot-encoding of data_categorical here

#join the two masked dataframes back together
data_joined = pd.concat([data_numeric, data_categorical], axis = 1)
```

```python
data_joined.head()
```

### 1.4.6.5 Extension To Imputation <a id='extension_imputation'></a>
Imputation is the standard approach, and it usually works well. However, imputed values may be systematically above or below their actual values (which weren't collected in the dataset). Or rows with missing values may be unique in some other way. In that case, your model would make better predictions by considering which values were originally missing.

![title](Images\Data_Imp_3.png)

In this approach, we impute the missing values, as before. And, additionally, for each column with missing entries in the original dataset, we add a new column that shows the location of the imputed entries.

In some cases, this will meaningfully improve results. In other cases, it doesn't help at all.

[Back to top](#Introduction)

```python
# make copy to avoid changing original data (when Imputing) and selecting only numeric columns to make our lives a little bit easy

melb_predictors_1 = df_1.drop(['Price'], axis=1)
X_1 = melb_predictors.select_dtypes(exclude=['object'])


# Get names of columns with missing values

cols_with_missing_1 = (col for col in X_1.columns 
                                 if X_1[col].isnull().any())
# make new columns indicating what will be imputed
for col in cols_with_missing_1:
    X_1[col + '_was_missing'] = X_1[col].isnull()

# Imputation

my_imputer = Imputer(missing_values=np.nan, strategy='mean', axis=0)
X_1 = pd.DataFrame(my_imputer.fit_transform(X_1), columns = X_1.columns)
X_1.head()
```

### 1.5 Implementation <a id='implementation'></a>

[Back to top](#Introduction)

```python
# Copy the oroginal dataset
data = df_1

# Select target
y = data.Price

# To keep things simple, we'll use only numerical predictors
melb_predictors = data.drop(['Price'], axis=1)
X = melb_predictors.select_dtypes(exclude=['object'])

# Divide data into training and validation subsets
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=10, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
# Get names of columns with missing values
cols_with_missing = [col for col in X_train.columns
                     if X_train[col].isnull().any()]

# Drop columns in training and validation data
reduced_X_train = X_train.drop(cols_with_missing, axis=1)
reduced_X_valid = X_valid.drop(cols_with_missing, axis=1)

print("MAE from Dropping columns with missing values:")
print(score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid))
```

```python
# Imputation
my_imputer_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
my_imputer_median = SimpleImputer(missing_values=np.nan, strategy='median')

#Mean
imputed_X_train_Mean = pd.DataFrame(my_imputer_mean.fit_transform(X_train))
imputed_X_valid_Mean = pd.DataFrame(my_imputer_mean.transform(X_valid))

#Median
imputed_X_train_Median = pd.DataFrame(my_imputer_median.fit_transform(X_train))
imputed_X_valid_Median = pd.DataFrame(my_imputer_median.transform(X_valid))

# Imputation removed column names; put them back
imputed_X_train_Median.columns = X_train.columns
imputed_X_valid_Median.columns = X_valid.columns

imputed_X_train_Mean.columns = X_train.columns
imputed_X_valid_Mean.columns = X_valid.columns

print("MAE from Imputation-Mean):")
print(score_dataset(imputed_X_train_Mean, imputed_X_valid_Mean, y_train, y_valid))

print("MAE from Imputation-Median):")
print(score_dataset(imputed_X_train_Median, imputed_X_valid_Median, y_train, y_valid))
```

```python
# Make copy to avoid changing original data (when imputing)
X_train_plus = X_train.copy()
X_valid_plus = X_valid.copy()

# Make new columns indicating what will be imputed
for col in cols_with_missing:
    X_train_plus[col + '_was_missing'] = X_train_plus[col].isnull()
    X_valid_plus[col + '_was_missing'] = X_valid_plus[col].isnull()

# Imputation
my_imputer_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
my_imputer_median = SimpleImputer(missing_values=np.nan, strategy='median')

imputed_X_train_plus_mean = pd.DataFrame(my_imputer_mean.fit_transform(X_train_plus))
imputed_X_train_plus_median = pd.DataFrame(my_imputer_median.fit_transform(X_train_plus))

imputed_X_valid_plus_mean = pd.DataFrame(my_imputer_mean.transform(X_valid_plus))
imputed_X_valid_plus_median = pd.DataFrame(my_imputer_median.transform(X_valid_plus))

# Imputation removed column names; put them back
#Mean
imputed_X_train_plus_mean.columns = X_train_plus.columns

#Median
imputed_X_valid_plus_median.columns = X_valid_plus.columns

print("MAE from An Extension to Imputation-Mean):")
print(score_dataset(imputed_X_train_plus_mean, imputed_X_valid_plus_mean, y_train, y_valid))

print("MAE from An Extension to Imputation-Median):")
print(score_dataset(imputed_X_train_plus_median, imputed_X_valid_plus_median, y_train, y_valid))
```

# <p style="text-align: center;">Conclusion<p><a id='Conclusion'></a>
As we can see, Approach 3(Extension to Imputation) performed slightly worse than Approach 2(Imputation).

So, why did imputation perform better than dropping the columns?

As is common, imputing missing values yielded better results, relative to when we simply dropped columns with missing values. The training data has 10864 rows and 12 columns, where three columns contain missing data. For each column, less than half of the entries are missing. Thus, dropping the columns removes a lot of useful information, and so it makes sense that imputation would perform better.

Also Central Tendency method of Median works slightly better than mean because of the distribution of data.

We have used various data cleaning and preprocessing techniques in this notebook. As listed below

- **Finding Missing Values:** Various techniques used to find missing values in the dataset'))
- **Finding Why are there Missing Values:** Various reasons suggesting why there would be missing values in the dataset'))
- **Handling Missing Values:** Suggesting techniques to handle the various missing values'))
- **Drop Missing Values:** The dirtiest technique, results in loss of quite a bit of data (reducing the sample size more than half in the example in this notebook)'))
- **Drop Columns with Missing Values:** Slightly better than dropping all rows with missing values, but in process of dropping columns we might lose some important predictors'))
- **Filling Missing Values(Automatically):** Blind filling the missing values with some set value like 0. Gives better results than techniques mentioned above, still not nearly as good.'))
- **Imputation:** By far the best technique to replace the missing dataset with a measure of central tendency (depending on the distribution). Gives the best result for the given dataset in this notebook'))
- **Extension to Imputation:** Extends Imputation with addition of extra columns which suggest where the data was imputed. May sometimes give better results than plain imputation. Not in the case of this dataset.
